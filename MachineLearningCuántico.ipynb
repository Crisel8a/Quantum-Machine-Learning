## Lanzar una Moneda Clásica

Los estados cuánticos pueden ser modelados como una generalizzación de las distribuciones clásicas de probabilidad y los cálculos cuánticos como una forma de transformarlas. Antes de iniciar con esto, repasemos algo de probablidad clásica, lancemos una moneda. Esta tendrá dos posibles resultados: {cara,cruz}. Asociemos la probabilidad $p_0$ a cara y $p_1$ a cruz, con $p_0, p_1 \in [0,1]$, tales que $p_0+p_1 = 1$, es decir $p_1 = 1 - p_0$
$$P(X=cara)=p_0, P(X=cruz)=p_1$$



## Entropía

Veamos ahora ottro concepto llamado entropía, la cual nos habla sobre la incertidumbre de una distribución de probabilidad. En una moneda clásica, el caso con mayor incertidumbre es justo al lanzar la moneda, pues no tenemos absolura idea de cuál será el resultado. 

## Vector Estocástico
Representamos ahora nuestras probabilidades en un vector columa, tal que $p_0, p_1 \geq 0$ y 
$\overrightarrow{p} = \begin{pmatrix} p_0 \\ p_1 \end{pmatrix}$. Notamos que la norma usual de este vector es 1, es decir, es un vector unitario.

¿cómo tranformamos distribuciones de probabilidad?, es decir, ¿como transformamos este vector?

## Transformando vectores estocásticos
Tomamos el vector estocástico $\overrightarrow{p}$ y queremos obtener un nuevo vectos estocástico $\overrightarrow{p'}$, una nueva distribución de probabilidad. Con esto, la transformación que apliquemos debe cumplir dichos requerimientos. Por supuesto aplicamos una matriz $M$ llamada *matriz estocástica*. Al aplicarla por la izquierda, esto es $M \cdot \overrightarrow{p}$, significa que la suma de sus columnas debe ser 1. 


## Estados Cuánticos
Así como los vectores estocásticos, podemos escribir los estados cuánticos como un vector columna con entradas $a_0, a_1 \in \mathbb{C}$
$$ \ket{\psi} = \begin{pmatrix} a_0 \\ a_1 \end{pmatrix} $$
tal que, con la norma usual de $\mathbb{C}$ se cumpla que $|| \ket{\psi}  || = \sqrt{|a_0|^2+|a_1|^2} = 1$

A los estados cuánticos con dos entradas las llamamos *qubits*.

## Superposición
No es más que la expansión de un estado cuántico en términos de una base.
$$\ket{\psi} = \begin{pmatrix} a_0 \\ a_1 \end{pmatrix} = a_0 \begin{pmatrix} 1 \\ 0 \end{pmatrix} + a_1 \begin{pmatrix} 0 \\ 1 \end{pmatrix} = a_0 \ket{0} + a_1\ket{1}$$
De donde, el estado 0 tiene una probabilidad de $|a_0|^2$ y el estado 1 con una probabilidad de $|a_1|^2$

Con la misma idea de usar la flecha sobre $p$ al denotar el vector, en estados cuánticos usamos la notación *ket*. El resultado no es más que una superposición del ket 0 y el ket 1.


## Esfera de Bloch
Si pensamos en una forma más geométrica para representar los estados, tenemos un vector de dimensión dos en un espacio complejo, tomando en cuenta la norma, podríamos entonces tener un obejto de dimensión tres representando el estado de un qubit. Aquí es donde la *esfera de Bloch* nos ayuda.  


El polo norte de la esfera, se identifica como el ket $0$, y el polo sur con el ket $1$. Es un poco inusual pues estos kets son ortogonales entre sí, mientras que en la esfera parecen estar en la misma línea, es importante tener en mente que la ortogonalidad es un poco diferente en esta esfera, recordando que trabajamos en un espacio complejo. Ahora, cada punto sobre la esfera representa el estado de un qubit, así, el estado de un qubit tiene toda una esfera de posibles estados. 

## Superposición
Algunas cosas que podemos hacer con estados cuánticos a comparación de estados clásicos es la superposición, veamos un ejemplo.

$$ M \ket{0} = \begin{pmatrix} \frac{1}{\sqrt{2}} && -\frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} && \frac{1}{\sqrt{2}} \end{pmatrix} \ket{0} = \frac{1}{\sqrt{2}} (\ket{0}+\ket{1}) $$

Notamos que el operador $M$ manda al ket $0$ a una superposición de los estados $\ket{0}$ y $\ket{1}$. Similarmente, si aplicamos $M$ al ket $1$, la única diferencia es que introduce un signo negativo en el ket $0$, esto es $\frac{1}{\sqrt{2}} (-\ket{0}+\ket{1})$
Lo que es interesante, es que si tomamos el resultado de $M \ket{0}$ y le aplicamos $M$ de nuevo, obtenemos
$$\frac{1}{\sqrt{2}}(\ket{0}+\ket{1}-\ket{0}+\ket{1}) $$
que es una superposición entre los dos estados aneteriormente obtenidos.

## Muchos Qubits
Tratamos ahora de construir distribuciones de probabilidad más grandes, estados más extensos compuestos de una muchos qubits, est lo logramos con el *producto tensorial* de dos estados cuánticos. Recordemos que, dados dos vectores $\ket{\psi} = \begin{pmatrix} a_0 \\ a_1 \end{pmatrix}$ y $\ket{\psi'} = \begin{pmatrix} b_0 \\ b_1 \end{pmatrix}$, su producto tensorial es
$$ \ket{\psi} \otimes \ket{\psi'} = \begin{pmatrix} a_0b_0 \\ a_0b_1 \\ a_1b_0 \\ a_1b_1 \end{pmatrix}$$

Tenemos entonces que


$$
    \ket{0} \otimes \ket{0} = \ket{00} = \myvector{1 \\ 0 \\ 0 \\ 0},
    ~~~~~~
    \ket{0} \otimes \ket{1} = \ket{01} = \myvector{0 \\ 1 \\ 0 \\ 0},
$$
$$
    ~~~~~~
    \ket{1} \otimes \ket{0} = \ket{10} = \myvector{0 \\ 0 \\ 1 \\ 0},
    ~~~~~~ 
    \ket{1} \otimes \ket{1} = \ket{11} = \myvector{0 \\ 0 \\ 0 \\ 1}.
$$


## Medición (pendiente)

Tratamos de medir variables aleatorias. Si luego de la medición tenemos al estado $0$ con probabilidad $|a_0|^2$, el estado observado será $\ket{0}$, esto significa que el estado colapsó al estado $\ket{0}$ lo cual se obtiene al destruir la superposición.
